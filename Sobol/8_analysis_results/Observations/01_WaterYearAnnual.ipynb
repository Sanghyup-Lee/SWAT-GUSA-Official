```python
import pandas as pd
from pathlib import Path

# === 1. 파일 경로 설정 (네가 말한 경로 그대로) ===========================
file_path = Path(r"Z:\WERL\Team\shlee\Sobol\8_analysis_results\Observations\Observation(Carlyle)csv.csv")

# === 2. 파일 읽기 ==========================================================
df = pd.read_csv(file_path)

# 컬럼 이름이 time, cms 라고 가정
df["time"] = pd.to_datetime(df["time"])

# === 3. Water Year 계산 (10월 1일 시작) ====================================
def water_year(dt):
    return dt.year + 1 if dt.month >= 10 else dt.year

df["water_year"] = df["time"].apply(water_year)

# === 4. Water Year 기준 Annual Max/Min/Avg 계산 =============================
annual_stats = df.groupby("water_year")["cms"].agg(
    annual_max_cms="max",
    annual_min_cms="min",
    annual_avg_cms="mean"
).reset_index()

# === 5. 결과 저장 ============================================================
output_path = Path(r"Z:\WERL\Team\shlee\Sobol\8_analysis_results\Observations\Annual_Stats_WY.csv")
annual_stats.to_csv(output_path, index=False)

print("Done! Saved to:", output_path)
print(annual_stats.head())
```
